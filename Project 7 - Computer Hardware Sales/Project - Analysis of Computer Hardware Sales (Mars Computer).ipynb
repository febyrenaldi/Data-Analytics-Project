# **Analysis in Hardware**

## **Hardware Sales**
The next step, I will look for total orders and total profit obtained from all types of hardware's name and hardware's categories :

```python
SELECT 
	hardware_name,
	hardware_category,
	COUNT(hardware_name) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY hardware_name, hardware_category
ORDER BY total_profit DESC;
```

```python
SELECT 
	hardware_category,
	COUNT(hardware_category) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY hardware_category
ORDER BY total_order DESC;
```

Based on the results above, it can be seen that hardware included in I-O (Input Output Hardware) is the type of category most ordered by customers.

However, if you look at the amount of profit obtained, it can be seen that Core Hardware provides a very high total profit compared to the other 3 categories.

This is because the prices of tools including core hardware have a high price value, so they can provide high profits too.

## **Sales by Hardware Category Each Year**

After analyzing sales in each customer type, the next step was to analyze sales based on the hardware category that had previously been analyzed

```python
SELECT  
	year, 
	hardware_category,
	COUNT(hardware_category) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY year, hardware_category
ORDER BY year DESC, total_order DESC;
```

Based on the query results above, it can be seen that the I-O Hardware category has a very high sales frequency when compared to other hardware categories. Just like the previous analysis, the highest total profit was generated by Core Hardware.

## **Top 5 Hardware Sales**

In the next step, I will try to calculate the top 5 products that are most sold by customers in 2022 and 2023.

```python
SELECT 
	hardware_name,
	COUNT(hardware_name) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY hardware_name
ORDER BY total_order DESC
LIMIT 5;
```

Based on the results above, it can be seen that the best-selling product is the 64 GB Flashdisk, followed by Headphones (Wired), Mouse (Wireless), Mechanical Keyboard, 32 GB Flashdisk. Flash disks are the best-selling product because the price is affordable and demand is very high.

# **Analysis in City**

The next step, I will try to focus on analyzing sales results. I will look for total orders and total profit obtained in each city all the time :

## **Sales by City and Year**

I also use ROLLUP to calculate total profit and total profit based on year and city

```python
SELECT 
	CASE WHEN year IS NULL THEN 'Total Profit All Year :'
		 ELSE year END AS year,
	CASE WHEN year = 2022 AND city IS NULL THEN 'Z. Total Profit 2022 :'
		 WHEN year = 2023 AND city IS NULL THEN 'Z. Total Profit 2023 :'
		 ELSE city END AS city,
	COUNT(city) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY ROLLUP(year, city)
ORDER BY year ASC, city ASC;
```

## Producst Sales in Top 6 City in Indonesia

Next, I will carry out calculations to calculate the best-selling products in 6 big cities in Indonesia (Jakarta, Bandung, Surabaya, Semarang, Yogyakarta, Palangkaraya):

```python
SELECT
	hardware_name,
	COUNT(hardware_name) AS total_order
FROM 'mars_computer_sales.csv'
WHERE city in (
			'Jakarta', 
			'Bandung',
			'Surabaya', 
			'Semarang', 
			'Yogyakarta', 
			'Palangkaraya')
GROUP BY hardware_name
ORDER BY total_order DESC
LIMIT 5;
```

Based on the query results above, the 64 GB Flashdisk is still the best-selling product in 5 big cities in Indonesia.

Next, I will calculate total orders for each city and year, which has the top 3 ranking relative to purchase frequency

```python
WITH CityYearTotalOrders AS (
    SELECT
        city,
        year,
        COUNT(*) AS total_orders
    FROM 'mars_computer_sales.csv'
    GROUP BY city, year
),
RankedOrders AS (
    SELECT
        city,
        year,
        total_orders,
        RANK() OVER (PARTITION BY year ORDER BY total_orders DESC) AS order_rank
    FROM CityYearTotalOrders
)
SELECT
    city,
    year,
    total_orders
FROM RankedOrders
WHERE order_rank IN (1,2,3);
```

I used the window function (RANK) to rank the total orders for each year.
Then, from the ranking results, this query retrieves the cities and years with the top 3 order rankings in each year.

# **Analysis in Customer Type**

## **Sales by Customer Type**

Next, I will focus on searching based on customer type, namely Subscriber and Non-Member. For the Subscriber type, the price is 5% cheaper compared to the Non Member type.

Now, I will try to calculate the total orders and total profit obtained from the two types of customers :

```python
SELECT 
	customer_type,
	COUNT(customer_type) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY customer_type
ORDER BY total_profit DESC;
```

Based on the table above, it can be seen that non-member customers are more dominant in making purchases and transactions compared to subscribers.

This can be a reference, that by reducing prices by 5% for subscribers, it is not an effective step for this shop to get a wide target market.

## **Sales by Customer Type Each Year**

After analyzing sales in each city, the next step was to analyze sales based on the type of customer that had previously been analyzed

```python
SELECT  
	year, 
	customer_type,
	COUNT(customer_type) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY year, customer_type
ORDER BY year DESC, total_order DESC;
```

Based on each year, it can be seen that sales in 2022 will be much higher compared to 2023. This is because sales data for 2023 is only updated until August or for 8 months. Meanwhile, in 2022, sales data will cover 1 full year (12 months).

Next, I will carry out calculations using the ROLLUP function, to find total sales based on year and customer type

```python
SELECT 
	CASE WHEN year IS NULL THEN 'Total Profit All Year :'
		 ELSE year END AS year,
	CASE WHEN year = 2022 AND customer_type IS NULL THEN 'Total Profit in 2022 :'
		 WHEN year = 2023 AND customer_type IS NULL THEN 'Total Profit in 2023 :'
		 ELSE customer_type END AS customer_type,
	COUNT(customer_type) AS total_order,
	SUM(profit_per_item) AS total_profit
FROM 'mars_computer_sales.csv'
GROUP BY ROLLUP(year, customer_type)
ORDER BY year ASC, customer_type ASC;
```

And now, I will calculate total orders for each combination of payment method and customer type :

```python
WITH PaymentTypeCustomerTypeTotalOrders AS (
    SELECT
        payment_method,
        customer_type,
        COUNT(*) AS total_orders
    FROM 'mars_computer_sales.csv'
    GROUP BY payment_method, customer_type
),
RankedOrders AS (
    SELECT
        payment_method,
        customer_type,
        total_orders,
        RANK() OVER (PARTITION BY payment_method, customer_type ORDER BY total_orders DESC) AS order_rank
    FROM PaymentTypeCustomerTypeTotalOrders
)
SELECT
    payment_method,
    customer_type,
    total_orders
FROM RankedOrders
WHERE order_rank = 1; --- Show only 1 in rank
```

I use the window function (RANK) to rank the total orders for each combination of payment method and customer type.

Then, from the ranking results, the query takes the combination of payment method and customer type with the top order ranking (rank 1).

This query combines various SQL concepts in one complex statement to perform in-depth analysis of the data by considering payment method, customer type, and total orders. Make sure you adapt this query according to the structure of the tables and columns in your dataset.

## **Reward for Customer Loyalty**

To increase customer loyalty and attention to purchasing products at the Mars Computer Online store, the company owner implemented the idea of giving rewards to 10 customers who are Subscribers, with M-Banking and E-Wallet payments and have the highest number of transactions.

The rewards given are as follows:

1. If a customer orders more than 7 times, the customer is entitled to get 1 unit of 64 GB Flashdisk for free.
2. If the customer orders more than 6 times, the customer is entitled to 1 unit of USB Mini Hub 3.0.
3. If the customer orders more than 5 times, the customer is entitled to 1 unit of Hard Disk Enclosure 2.0.
All costs for sending rewards to customers are borne by the shop.

```python
SELECT 
	customer_id,
	city,
	COUNT(customer_id) AS total_order,
    RANK() OVER (ORDER BY COUNT(customer_id) DESC) AS order_rank,
	CASE WHEN total_order > 7 THEN 'Flashdisk 64 GB'
		 WHEN total_order > 6 THEN 'USB Mini Hub 3.0'
		 WHEN total_order > 5 THEN 'Harddisk Enclosure 2.0'
		 END AS reward
FROM 'mars_computer_sales.csv'
WHERE customer_type = 'Subscriber' AND payment_method IN('M-Banking','E-Wallet')
GROUP BY customer_id, gender, city
ORDER BY total_order DESC
LIMIT 10;
```
